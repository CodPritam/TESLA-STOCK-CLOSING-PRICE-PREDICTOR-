## Let's know about the data....
Date: This column indicates the date for which the stock price is recorded. It helps establish the chronological order of the data points.

Open: The opening price of Tesla’s stock on a particular trading day. This is the price at which the stock starts trading when the market opens.

High: The highest price at which Tesla’s stock traded during the trading day.

Low: The lowest price at which Tesla’s stock traded during the trading day.

Close: The closing price of Tesla’s stock on a particular trading day. This is the price at which the stock ends trading when the market closes.

Volume: The total number of shares traded of Tesla’s stock on that particular trading day. It indicates the level of investor interest and activity in the stock.

Adjusted Close: Sometimes, this column is provided to account for corporate actions (like dividends or stock splits) that affect the stock price. It adjusts the closing price accordingly.

What to Predict:

When predicting Tesla stock prices, the primary target is usually the closing price. This is because the closing price reflects the final value of the stock for that trading day and is often used as a benchmark for performance measurement.

```python
#Now I will apply random forest model on the data set ...let's see it will give us output...
#let's first split data for both training and testing purpose ..
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
X = data[['Open', 'High', 'Low', 'Volume']]
y = data['Close']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# Train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=1)
rf_model.fit(X_train, y_train)

# Make predictions
rf_predictions = rf_model.predict(X_test)
plt.plot(y_test.values, label='Actual Prices')
plt.plot(rf_predictions, label='Predicted Prices')
plt.title('Random Forest Model Predictions')
plt.legend()
plt.show()
```

## We get from this plot that our predictor model is predicting absolutely right ...but it lacks in predicting  some higher values...

## Let's check it's accuracy report  report ..............

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Calculate metrics
mae = mean_absolute_error(y_test,rf_predictions)
mse = mean_squared_error(y_test, rf_predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, rf_predictions)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

So our model is godd but can be better .....
An R² value of 0.72 means that 72% of the variance in the dependent variable (stock prices) is predictable from the independent variables (features). This indicates a relatively good fit of the model to the data.

## So here we have good model and no need of performance tuning .....so let's use this model for random nnumbers for predicting the closing value..

```python
#Let's first save the model ..................
import joblib
# Save the trained model to a .pkl file
filename = 'random_forest_regressor_model.pkl'
joblib.dump(rf_model, filename)

print(f"Model saved as {filename}")
```

```python
#Let's create our API ........
from flask import Flask, request, render_template
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import joblib

# Load the trained model
model = joblib.load('random_forest_regressor_model.pkl')  # Replace with your actual model file

app = Flask(__name__)

# Home route
@app.route('/')
def home():
    return render_template('index.html')

# Prediction route
@app.route('/predict', methods=['POST'])
def predict():
    # Get user input from form
    opening = float(request.form['opening'])
    high = float(request.form['high'])
    low = float(request.form['low'])
    volume = float(request.form['volume'])
    
    # Prepare input data for prediction
    input_data = np.array([[opening, high, low, volume]])
    
    # Predict the close value
    predicted_close = model.predict(input_data)
    
    # Render prediction result template with predicted close value
    return render_template('result.html', opening=opening, high=high, low=low, volume=volume, predicted_close=predicted_close[0])

if __name__ == '__main__':
    app.run(debug=True)
```
